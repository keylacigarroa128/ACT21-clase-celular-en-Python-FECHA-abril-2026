```python
class Celular:
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    def encender(self):
        print(f"El {self.modelo} se ha encendido.")
        self.encendido = True
        print(f"El celular está encendido: {self.encendido}")

    def mostrar_info(self):
        print(f"Celular: {self.marca} {self.modelo}, Precio: ${self.precio}")
# Instanciación
celular1 = Celular("Samsung", "Galaxy S23", 800)
celular2 = Celular("Apple", "iPhone 15", 1000)
#llamada a las funciones para cada objeto celular 1 y celular 2
print("celular 1:")
celular1.mostrar_info()
celular1.encender()

print("celular 2:")
celular2.encender()
celular2.mostrar_info()

print("Programa terminado")
print("Creador de Software por Keyla Paola Cigarroa Palacios NL = 12")
```
